```python
#Univariate Analysis

#Price distribution

#Identifying Outliers

sns.boxplot(data = data_ab,x = "price")
```

```python
#Handling outliers

data = data_ab[data_ab['price'] < 1500]
```

```python
#After handling outliers

sns.boxplot(data = data,x = "price")
```

```python
#Price distribution plotting

plt.figure(figsize=(10,5))
sns.histplot(data = data['price'],bins=100)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
#Availability_365 Distribution

plt.figure(figsize=(10,5))
sns.histplot(data = data['availability_365'])
plt.xlabel('Availability')
plt.ylabel('Frequency')
plt.title('Availability_365 Distribution')
plt.show()
```

```python
#Avg price by neighbourhood_group

data.groupby(by='neighbourhood_group')['price'].mean()
```

```python
#Feature Engineering

#Add a new column to data --> 'price_per_bed'

data['price_per_bed'] = data['price']/data['beds']
```

```python
#Avg price per bed by neighbourhood_group

data.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
#Bivariate Analysis

#Price dependency on Neighbourhood_group

sns.barplot(data = data,x = 'neighbourhood_group',y = 'price',hue = 'room_type')
plt.xlabel('Neighbourhood_group')
plt.ylabel('Price')
plt.title('Price dependency on Neighbourhood_group')
plt.show()
```

```python
#Locality and Review dependency on Price

sns.scatterplot(data = data,x = 'number_of_reviews',y = 'price',hue = 'neighbourhood_group')
plt.xlabel('Number_of_reviews')
plt.ylabel('Price')
plt.title('Locality and Review dependency on Price')
plt.show()
```

```python
#Relationship b/w Price,Minimum_nights,Number_of_reviews,Availability_365

sns.pairplot(data = data,vars = ['price','minimum_nights','number_of_reviews','availability_365',],hue = 'room_type')
plt.show()
```

```python
#Geographical Distribution

sns.scatterplot(data = data,x = 'latitude',y = 'longitude',hue = 'room_type')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geographical Distribution')
plt.show()
```

```python
#Correlation

co_rel = data[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
sns.heatmap(data = co_rel,annot = True)
plt.show()
```
